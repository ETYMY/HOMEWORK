```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error,mean_absolute_percentage_error
```

```python
# Load the data
mv=["#"]
df = pd.read_csv('breast-cancer-dataset.csv',na_values=mv)
print(df.to_string())
```

```python
df = df.dropna(subset=['S/N', 'Year','Menopause','Tumor Size (cm)','Inv-Nodes','Breast','Metastasis','Breast Quadrant','History','Diagnosis Result','Age'])  # Drop rows where there are NaN
```

```python
print(df.to_string())
```

## KETEGORIK VERILERI DONUSUMU

```python
df=pd.get_dummies(df,columns=["Breast","Breast Quadrant","Diagnosis Result"])
```

```python
print(df)
```

```python
df=df.astype(int)
```

```python
print(df)
```

## Z-skoru hesaplanmasi

```python
from scipy import stats
import numpy as np

z=np.abs(stats.zscore(df))
print(z)
```

```python
print(np.where (z>3))
```

```python
selected=df.iloc[np.where (z>3)[0]].index
print(selected)
```

```python
 dfN=df.drop(selected,axis=0,inplace=True)
```

```python
dfN=df
print(dfN.shape)
```

## Normalisasyon

```python
 dfN=(dfN-dfN.min())/(dfN.max()-dfN.min())
print(dfN.head())
```

```python
dfN = dfN.dropna(subset=['S/N', 'Year','Menopause','Tumor Size (cm)','Inv-Nodes','Breast_Left','Breast_Right','Metastasis','Breast Quadrant_Lower inner','Breast Quadrant_Lower outer','Breast Quadrant_Upper inner','Breast Quadrant_Upper outer','Breast Quadrant_Upper outer','History','Diagnosis Result_Benign','Diagnosis Result_Malignant','Age'])
```

## veri setini Ã¶zelliklere (x) ve hedefe (y) ayÄ±rma

```python
y=df.loc[:,["Diagnosis Result_Benign","Diagnosis Result_Malignant"]].values
x=df.drop(["Diagnosis Result_Benign","Diagnosis Result_Malignant"],axis=1)
```

## verileri eÄŸitim ve test kÃ¼mesine bÃ¶lme

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
# Ã§oklu doÄŸrusal regresyon modelinin eÄŸitimi

mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
# tahmimi yaplmasi
mlr_pred=mlr.predict(x_test)
```

```python
print(mlr.intercept_,mlr.coef_)
```

## metriklerin hesaplanmasÄ±

```python
print("mlr",r2_score(y_test,mlr_pred),mean_absolute_percentage_error(y_test,mlr_pred),mean_absolute_error(y_test,mlr_pred),mean_squared_error(y_test,mlr_pred))
```

## Polynomial linear regression

```python
from sklearn.preprocessing import PolynomialFeatures
plr=PolynomialFeatures(degree=3)
x_pol=plr.fit_transform(x)
lr_pol=LinearRegression()
lr_pol.fit(x_pol,y)
print(lr_pol.intercept_,lr_pol.coef_)
```

## Decision Tree

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt_pred=dt.predict(x_test)
```

```python
print("dt",r2_score(y_test,dt_pred),mean_absolute_percentage_error(y_test,dt_pred),mean_squared_error(y_test,dt_pred),mean_absolute_error(y_test,dt_pred))
```

## Random Forest

```python
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
```

```python
print("rf",r2_score(y_test,rf_pred),mean_absolute_percentage_error(y_test,rf_pred),mean_squared_error(y_test,rf_pred),mean_absolute_error(y_test,rf_pred))
```
